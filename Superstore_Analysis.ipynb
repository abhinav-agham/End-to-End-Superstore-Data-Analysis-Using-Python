# grouping data by category 

```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
```

```python
# Sorting to make the chart look professional
category_sales = category_sales.sort_values(by='Sales', ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=category_sales, palette='viridis')

# Adding labels and title
plt.title('Total Sales by Category', fontsize=15)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Sales (USD)', fontsize=12)

plt.show()
```

### 📊 Key Insights from Category Sales
* *Top Performer:* The **Technology** category leads the sales charts, indicating it is the primary revenue driver for the Superstore.
* *Stable Demand:* **Furniture** follows closely, showing consistent high-value transactions.
* *Growth Opportunity:* **Office Supplies** shows the lowest sales volume, suggesting a need for better promotional strategies or product bundling in this segment

# Check Total Profit by Region

```python
# Grouping by Region for Profit analysis
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False).reset_index()

# Visualization
plt.figure(figsize=(8, 5))
sns.barplot(x='Region', y='Profit', data=region_profit, hue='Region', palette='rocket', legend=False)

plt.title('Total Profit by Region', fontsize=15)
plt.show()
```

### 🌍 Key Insights: Regional Profitability
* *Leading Market:* The **West Region** is the most profitable area, suggesting a strong customer base and efficient operations.
* *Operational Concern:* Despite having decent sales, some regions like Central might show lower profit margins. This could be due to high logistics costs or heavy discounting.
* *Strategic Focus:* Maintaining the momentum in the West while investigating the causes of lower profits in other regions should be a priority.


# Check Annual Sales Trend (Yearly)

```python
# Order Date converting datetime format  
df['Order Date'] = pd.to_datetime(df['Order Date'])

# create new 'Year' column 
df['Year'] = df['Order Date'].dt.year

# Year wise grouping of sales
yearly_sales = df.groupby('Year')['Sales'].sum().reset_index()

# Line Chart 
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Sales', data=yearly_sales, marker='o', color='b', linewidth=2.5)

# Labels and Title
plt.title('Annual Sales Trend (Yearly)', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6) 

plt.show()
```

### 📈 Business Growth Trend
* *Consistent Growth:* The store has shown a steady upward trend in sales year-over-year, indicating successful business expansion.
* *Peak Performance:* The final year in the dataset shows the highest revenue, suggesting that marketing and operational strategies are yielding positive results.
* *Forecast Hint:* Based on this growth trajectory, we can expect further increases in the coming year, provided market conditions remain stable.

# Checking impact of Discount on Profitability

```python
plt.figure(figsize=(12, 7))

# Background Trend Line (Red line showing the drop)
sns.regplot(x='Discount', y='Profit', data=df, scatter=False, color='red', line_kws={"linewidth":2})

# Scatter plot with colors (Green for Profit, Red for Loss)
# 'c' argument defines color based on profit
plt.scatter(df['Discount'], df['Profit'], 
            c=(df['Profit'] > 0), cmap='RdYlGn', alpha=0.5, s=30)

# Zero line for clear distinction
plt.axhline(0, color='black', linestyle='--', alpha=0.6)

plt.title('Impact of Discount on Profitability', fontsize=15, fontweight='bold')
plt.xlabel('Discount Percentage (0 to 1)', fontsize=12)
plt.ylabel('Profit / Loss (USD)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)

plt.show()
```

### 📉 Correlation Analysis: Discount vs. Profit
* *Negative Impact:* There is a clear trend showing that as *Discount increases beyond 0.2 (20%)*, the business starts incurring significant losses.
* *Loss Zones:* Most of the deep losses negative profit are clustered at discount levels of 0.5 or higher.
* *Strategic Recommendation:* The company should reconsider its discounting policy. High discounts are driving sales volume but destroying the bottom line profit.


# top 10 most profitable products list 

```python
# Data preparation
top_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10).reset_index()

# Visualization
plt.figure(figsize=(12, 8))

sns.barplot(
    y='Product Name', 
    x='Profit', 
    data=top_products, 
    hue='Product Name', 
    palette='magma', 
    legend=False
)

plt.title('Top 10 Most Profitable Products', fontsize=16, fontweight='bold')
plt.xlabel('Total Profit (USD)', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.tight_layout()
plt.show()
```

### 🏆 Key Insights: Top Performing Products
* *Profit Leaders:* The chart highlights that **Canon imageCLASS 2200 Advanced Copier** is the single largest profit contributor for the store.
* *Technology Dominance:* High-value technical equipment (Copiers and Printers) occupies most of the top spots, proving that the *Technology category* is the backbone of our profit.
* *Business Strategy:* Since these 10 products drive a massive portion of the total profit, they should be prioritized for inventory management and premium customer support.
